```python
import numpy as np
import matplotlib.pyplot as plt
import os
import pandas as pd

%matplotlib inline

PATH = os.path.abspath('..')
```

```python
a40 = pd.read_csv(f'{PATH}/results/predictions/best_a40A-34.csv')
```

```python
plt.figure(figsize=(6.2, 5), dpi=150)
plt.hist2d(a40.fgas_true, a40.fgas_pred, cmap='Blues', range=[(-1.5, 1.5), (-1.5, 1.5)], bins=30);
plt.plot([-2, 2], [-2, 2], c='black', ls='-', lw=2)
plt.gca().set_aspect('equal')
plt.colorbar();
plt.xlabel(r'$\mathcal{M}_{\rm true}$', fontsize=12)
plt.ylabel(r'$\mathcal{M}_{\rm pred}$', fontsize=12);
plt.xticks(np.arange(-1.5, 2., .5))
plt.yticks(np.arange(-1.5, 2., .5));

a40_rmse = np.sqrt(np.mean(np.power(a40.fgas_true - a40.fgas_pred, 2)))
plt.text(-1.3, 1.1, r'$\alpha$.40' +'\n' + f'RMSE = {a40_rmse:.2f} dex', fontsize=12);

plt.savefig(f'{PATH}/results/figures/results-hist_a40A.pdf')
```

```python
xg = pd.read_csv(f'{PATH}/results/predictions/best_xGASS-34.csv')
```

```python
plt.figure(figsize=(6.2, 5), dpi=150)
plt.hist2d(xg.fgas_true, xg.fgas_pred, cmap='Blues', range=[(-2.25, 0.75), (-2.25, 0.75)], bins=30);
plt.plot([-3, 2], [-3, 2], c='black', ls='-', lw=2)
plt.gca().set_aspect('equal')
plt.colorbar();
plt.xlabel(r'$\mathcal{M}_{\rm true}$', fontsize=12)
plt.ylabel(r'$\mathcal{M}_{\rm pred}$', fontsize=12);
plt.xticks(np.arange(-2.25, 1.0, .5))
plt.yticks(np.arange(-2.25, 1.0, .5));

xg_rmse = np.sqrt(np.mean(np.power(xg.fgas_true - xg.fgas_pred, 2)))
plt.text(-2.05, 0.35, r'xGASS' +'\n' + f'RMSE = {xg_rmse:.2f} dex', fontsize=12);

plt.savefig(f'{PATH}/results/figures/results-hist_xGASS.pdf')
```

# Combined into one

```python
a40 = pd.read_csv(f'{PATH}/results/predictions/best_a40A-34.csv')
xg = pd.read_csv(f'{PATH}/results/predictions/best_xGASS-34.csv')

a40_rmse = np.sqrt(np.mean(np.power(a40.fgas_true - a40.fgas_pred, 2)))
xg_rmse = np.sqrt(np.mean(np.power(xg.fgas_true - xg.fgas_pred, 2)))
```

```python
plt.figure(figsize=(6.2, 5), dpi=150)
plt.hist2d(a40.fgas_true, a40.fgas_pred, cmap='YlGnBu', range=[(-2.5, 1.5), (-2.5, 1.5)], bins=40)
cb = plt.colorbar();
cb.ax.set_title(r'$\alpha$.40A' +'\n' + f'RMSE = {a40_rmse:.2f} dex')

plt.scatter(xg.fgas_true, xg.fgas_pred, marker='.', s=10, c='C3',
            label='xGASS \n' + f'RMSE = {xg_rmse:.2f} dex');

plt.plot([-2.5, 2], [-2.5, 2], c='k', ls='--', lw=2)
plt.gca().set_aspect('equal')

plt.xlabel(r'$\mathcal{M}_{\rm true}$', fontsize=12)
plt.ylabel(r'$\mathcal{M}_{\rm pred}$', fontsize=12);
plt.xticks(np.arange(-2.5, 2., .5))
plt.yticks(np.arange(-2.5, 2., .5));

plt.legend(fontsize=12)

plt.savefig(f'{PATH}/results/figures/results-hist_a40A')
```
