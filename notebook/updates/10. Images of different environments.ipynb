```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

PATH = Path('../..').resolve()

%matplotlib inline
```

```python
def load_image(idx, df, ax=None, annotate=True, legacy=False):
    """Helper function to load and annotate image"""
    
    if not ax:
        fig, ax = plt.subplots(1, 1, figsize=(4, 4))
        
    d_5 = df.iloc[idx]
    name = df.index[idx]
    
    # get image
    if legacy:
        image = plt.imread(f'{PATH}/images-legacy/{name}.jpg')
    else:
        image = plt.imread(f'{PATH}/images-OC/{name}.jpg')
    ax.imshow(image)
    ax.axis('off')
        
    # label
    if annotate:
        fontdict = dict(color='white', size=18)
        ax.text(0.5, 0.9, f'AGC {name}', fontdict=fontdict, transform=ax.transAxes, ha='center', va='center')
```

# SDSS imaging

```python
df = (
    pd.read_csv(f'{PATH}/data/a40-SDSS_gas-frac.csv', usecols=(1,))
        .join(pd.read_csv(f'{PATH}/data/a.40-delta_5.csv', index_col=1, usecols=(1,2)), on='AGCNr')
    .set_index('AGCNr')
    .dropna()
)
```

```python
target_ld5 = -0.5
df = df.iloc[np.argsort((df.delta_5.apply(np.log10) - target_ld5)**2)]

fig, axes = plt.subplots(1, 5, figsize=(20, 4))
for idx in range(5):
    ax = axes.flat[idx]
    load_image(idx, df, ax=ax, legacy=False)
    
fig.tight_layout()

fig.subplots_adjust(wspace=0.02, hspace=0.02, left=0.02, right=1)
fig.text(0.005, 0.5, r'log(1+$\delta_5$) = ${:.1f}$'.format(target_ld5), rotation=90, fontsize=24, va='center');

plt.savefig(f'{PATH}/results/figures-updated/environment_examples_a.pdf')
```
